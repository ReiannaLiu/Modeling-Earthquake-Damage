```python
import pandas as pd

# Load the datasets
X_train = pd.read_csv('/content/drive/MyDrive/Stats 5243 Project I/X_train.csv')
y_train = pd.read_csv('/content/drive/MyDrive/Stats 5243 Project I/y_train.csv')
X_test = pd.read_csv('/content/drive/MyDrive/Stats 5243 Project I/X_test.csv')
y_test = pd.read_csv('/content/drive/MyDrive/Stats 5243 Project I/y_test.csv')

X_train.head(), y_train.head(), X_test.head(), y_test.head()
print(type(y_test))
```

```python
print(f'X_train shape: {X_train.shape}')
print(f'y_train shape: {y_train.shape}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score

# Initialize the Nearest Neighbors model
knn = KNeighborsClassifier()

# Fit the model on the corrected training data
knn.fit(X_train, y_train)

```

```python
# Make predictions on the test set
y_pred = knn.predict(X_test)
y_pred
```

```python
# Convert the 'y_pred' to a data frame
y_pred_df = pd.DataFrame(y_pred)
y_pred_df
```

```python

# Convert the 'y_test' and 'y_pred_df' to the same format
y_test_series = y_test['damage_grade']
y_pred_series = y_pred_df[1]
```

```python

# Calculate the weighted F1 score
weighted_f1 = f1_score(y_test_series, y_pred_series, average='weighted')

print(f'Weighted F1 Score: {weighted_f1}')
```

```python
from sklearn.metrics import accuracy_score

# Assuming y_test and y_pred are already defined from the KNN model predictions
# Calculate accuracy
accuracy = accuracy_score(y_test_series, y_pred_series)

accuracy
```

```python
from sklearn.metrics import confusion_matrix

# Calculate confusion matrix for KNN model
conf_matrix_knn = confusion_matrix(y_test_series, y_pred_series)

conf_matrix_knn
```

```python
from sklearn.metrics import classification_report

# Generate a classification report
report = classification_report(y_test_series, y_pred_series, output_dict=False)
print(report)
```
